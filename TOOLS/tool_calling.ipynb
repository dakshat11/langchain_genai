```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
# create tool

@tool
def multipy(a: int, b: int) -> int:
    """Given 2 numebers a and b this tool returns thier product"""
    return a * b
```

```python
print(multipy.invoke({'a' : 5 , 'b' : 3}))
```

```python
llm = ChatOpenAI()
```

```python
llm_with_tool = llm.bind_tools([multipy])
```

```python
llm_with_tool.invoke('hi how are you')
```

```python
query = HumanMessage('can you multiply 12 and 13 for me')

message = [query]
```

```python
result = llm_with_tool.invoke(message) #.tool_calls 
```

```python
result.tool_calls[0]['args']
```
